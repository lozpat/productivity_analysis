```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
x = np.random.choice(np.arange("2010-01-01", "2020-01-01", dtype="datetime64[D]"), 200)
y = np.random.normal(size=len(x))

fig, ax = plt.subplots()
ax.xaxis.update_units(x)
sns.regplot(x=ax.xaxis.convert_units(x), y=y)
```

```python
x = np.linspace(0,1,1000)
f = 2

y = np.sin(2*np.pi*f*x) #+ np.random.normal(scale=0.1, size=len(x))
#plt.plot(x,y)

poly = np.polyfit(x, y, deg=12)

fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(x, y, label='data')
ax2.plot((np.polyval(poly, x) ), label='fit')
ax1.legend()
ax2.legend()
```
